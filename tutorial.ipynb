### - pokemon.csv 변수
- \# : 포켓몬 ID
- Name : 포켓몬 이름
- Type 1 : 포켓몬 타입
- Type 2 : 포켓몬이 가진 두번째 타입 (필수는 x)
- HP : 포켓몬 체력
- Attack : 포켓몬 기본 공격력
- Defense : 포켓몬의 기본 방어력
- Sp.Atk : 포켓몬의 스페셜 공격력
- Sp.Def : 포켓몬의 스페셜 방어력
- Speed : 포켓몬의 속도 ( 더 빠른 포켓몬이 각 라운드에 시작할 때 먼저 공격함, 속도가 같으면 first pokemon이 먼저 공격)
- Generation : 몇세대 포켓몬인지
- Legendary : 전설의 포켓몬 여부

```python
countByFirst = combat.groupby('First_pokemon').count()
```

```python
countBySecond = combat.groupby('Second_pokemon').count()
```

```python
numberOfWins = combat.groupby('Winner').count()
numberOfWins # 이긴횟수
```

```python
losing_pokemon_index = np.setdiff1d(countByFirst.index.values, numberOfWins.index.values)-1
```

```python
countByFirst.index.values
```

```python
numberOfWins.index.values
```

```python
numberOfWins['Total Fights'] = countByFirst.Winner + countBySecond.Winner
```

```python
numberOfWins['Win Percentage'] = numberOfWins.First_pokemon/numberOfWins['Total Fights']
```

```python
countByFirst
```

```python
countBySecond.Winner
```

```python
numberOfWins
```

```python
total = pd.merge(pokemon, numberOfWins, right_index = True, left_on = '#')
total.sort_values('Win Percentage',ascending=False)
```

```python
pokemon['Type 1'].unique()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(total['total_power'],total['Win Percentage'])
plt.title('총력(공격력+방어력)에 따른 승률')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(total['Attack'],total['Win Percentage'])
plt.title('공격력에 따른 승률')
plt.show()
```
