```python
import numpy as np
import pandas as pd
```

```python
combats = pd.read_csv('combats.csv')
pokemon = pd.read_csv('pokemon.csv')
poke = pd.read_csv('pokemon_preprocessed.csv')
```

```python
weaknesses = pd.read_json("weaknesses_of_each_pokemon_type.json")
weaknesses.set_index('name',inplace=True)
weaknesses.head()
```

```python
pokemon.head()
```

```python
pokemon['Type 2'] = pokemon['Type 2'].fillna('None')
pokemon['Name'] = pokemon['Name'].fillna('Primeape')
```

```python
pokemon['Type 1'] = pokemon['Type 1'].replace('Fighting', 'Fight')
pokemon['Type 2'] = pokemon['Type 2'].replace('Fighting', 'Fight')
```

```python
pokemon['Legendary'] = pokemon['Legendary'].map({False: 0, True:1})
```

```python
pokemon.loc[pokemon['Name'].isnull()==True]
```

```python
pokemon.set_index('#', inplace=True)
```

```python
first_pokemon = combats.join(pokemon, on='First_pokemon',how='left')
```

```python
second_pokemon = combats.join(pokemon, on='Second_pokemon', how = 'left')
```

```python
first_pokemon.drop(['First_pokemon', 'Second_pokemon', 'Winner', 'Name', 'Generation', 'Legendary'], axis=1, inplace=True)
second_pokemon.drop(['First_pokemon', 'Second_pokemon', 'Winner', 'Name', 'Generation', 'Legendary'], axis=1, inplace=True)
```

```python
pokemon.head()
```

```python
first_pokemon_num = first_pokemon.iloc[:,2:8]
second_pokemon_num = second_pokemon.iloc[:,2:8]
first_pokemon_num.head()
```

```python
pokemon_num = second_pokemon_num - first_pokemon_num
pokemon_num.head()
```

```python
combats.Winner[combats.Winner == combats.First_pokemon] = 0
combats.Winner[combats.Winner == combats.Second_pokemon] = 1

combats.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
forest = RandomForestClassifier()
train_set, test_set = train_test_split(pokemon_num, test_size=0.4)
test_set, valid_set = train_test_split(test_set, test_size=0.5)
```

```python
forest.fit(X=train_set.iloc[:,1:], y=train_set.iloc[:,0])
```

```python
forest.score(X=test_set.iloc[:,1:], y=test_set.iloc[:,0])
```
