```python
pokemon = pd.read_csv('./pokemon_df_diff.csv')
pokemon.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1, inplace= True)
```

```python
pokemon
```

```python
cat_cols = ['#_x', 'First_pokemon', 'Second_pokemon', 
            'Legendary_first', 'Legendary_second']
conti_cols = ['HP_first', 'Attack_first',
              'Defense_first', 'Sp. Atk_first', 'Sp. Def_first', 'Speed_first',
              'Generation_first', 
              'HP_second', 'Attack_second', 'Defense_second',
              'Sp. Atk_second', 'Sp. Def_second', 'Speed_second', 'Generation_second']

y = ['Winner_mark']
```

```python
# OHE 적용 안할 때,
# X = pokemon[cat_cols + conti_cols]
# y = pokemon[y]
```

### OHE

```python
poke_dummies = pd.get_dummies(pokemon, columns=cat_cols)
poke_dummies
```

### X, y split

```python
if poke_dummies:
    X = poke_dummies[poke_dummies.columns.difference(y)]
    y = poke_dummies[y]
    assert poke_dummies.shape[1] == X.shape[1] + y.shape[1]
else:
    X = pokemon[cat_cols + conti_cols]
    y = pokemon[y]
    assert len(pokemon.columns) == len (set(pokemon.columns))
```

### Train/test split

```python
# train, test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=1)

# train / validation split
# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, stratify=y_train, random_state=1)

y_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
y_train = np.ravel (y_train, order = "c")
y_train.shape
```

```python
assert len(X_train.columns) ==len(set(X_train.columns))
```

### SVC-proba

```python
pipeline_svc = make_pipeline(StandardScaler(), SVC(kernel='linear', probability=True),verbose=True)
pipeline_svc.steps # class 이름 소문자로 변형하여 만들어준다.
```

```python
%%time
pipeline_svc.fit(X_train, y_train)
```

```python
accuracy_score(y_train, pipeline_svc.predict(X_train)), accuracy_score(y_test, pipeline_svc.predict(X_test))
```
