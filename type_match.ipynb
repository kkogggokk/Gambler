```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pokemon_df = pd.read_csv('./pokemon.csv')
combats_df = pd.read_csv('./combats.csv')
pokemon_df_diff = pd.read_csv('./dataset/pokemon_df_diff.csv')
```

```python
type2_nan_df= pokemon_df_diff[['#_x','First_pokemon','Type 1_first','Type 2_first','Second_pokemon','Type 1_second','Type 2_second','Winner_mark']]
type2_nan_df
```

```python
#타입2 0인 non값만 추출
first_type2_nan = type2_nan_df[(type2_nan_df['Type 2_first']==0)|(type2_nan_df['Type 2_second']==0)]
```

```python
first_type2_nan
```

```python
#type2컬럼들 삭제
sample0_df = first_type2_nan.drop(columns=["Type 2_first",'Type 2_second'])
sample0_df
```

```python
#first_pokemon type(15) vs second_pokemon type(13) 매치내역 조회
sample1 = sample0_df[(sample0_df['Type 1_first']==15)&(sample0_df['Type 1_second']==13)]
sample1
```

```python
#first_pokemon (9) vs second_pokemon (13) 매치내역 조회
#매치내역 없음
sample2 = sample0_df[(sample0_df['Type 1_first']==13)&(sample0_df['Type 1_second']==15)]
sample2
```

```python
#데이터 병합
#first_pokemon (물(11)) vs second_pokemon 아이스(14)) 매치내역
#first_pokemon 아이스(14) vs second_pokemon 물(11) 매치내역
sample_result = pd.concat([sample1,sample2])
sample_result
```

```python
# type(15) 이긴 횟수 = first_pokemon(15)가 이긴 횟수 + first_pokemon(13)이 진 횟수(=15가 이긴판)
# type(15) 승률 = (type(15)이긴 횟수 / 전체 횟수) *100
# 승률 = 승수 ÷ (승수+패수))×100

# first_pokemon Type(15)가 이긴 횟수
sample4 = sample_result[(sample_result['Type 1_first']==15)&(sample_result['Winner_mark']==1)]
sample4_l = len(sample4)
sample4_l
```

```python
# first_pokemon Type(13)이 진 횟수
sample5 = sample_result[(sample_result['Type 1_first']==13)&(sample_result['Winner_mark']==2)]
sample5_l = len(sample5)
sample5_l
```

```python
# type(15)가 이긴 전체 횟수
sample_l = sample4_l + sample5_l
sample_l
```

```python
# 전체 경기 횟수
sample_result_l = len(sample_result)
sample_result_l
```

```python
# type(15)의 승률 result= (sample4_l/sample_result_l)*100
result = (sample_l/sample_result_l)*100
result
```
